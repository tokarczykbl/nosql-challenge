5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
establishments.update_many({},[{'$set': {'geocode.latitude': { '$toDouble': '$geocode.latitude' },'geocode.longitude': { '$toDouble': '$geocode.longitude' }}}])
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
establishments.update_many({},[{'$set': {'RatingValue': { '$toInt': '$RatingValue'}}}])
```

```python
# Check that the coordinates and rating value are now numbers
# Creating empty list to hold data types
ratings = []
lats = []
longs = []

# looping through documents in establishment collection to find the data types of RatingValue, Longitude, and Latitude
for document in establishments.find():
    if 'RatingValue' in document:
        ratings.append(type(document['RatingValue']))
    if 'geocode' in document and 'latitude' in document['geocode']:
        lats.append(type(document['geocode']['latitude']))
    if 'geocode' in document and 'longitude' in document['geocode']:
        longs.append(type(document['geocode']['longitude']))

# Creating sets in order to get the unique values in the lists
unqiue_rating = set(ratings)
unqiue_lats = set(lats)
unqiue_longs = set(longs)

# Printing results
print(f'RatingValue data types {unqiue_rating}')
print(f'Latitude data types {unqiue_lats}')
print(f'Longitude data types {unqiue_longs}')
```
